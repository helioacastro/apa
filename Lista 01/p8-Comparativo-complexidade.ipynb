```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Definir o intervalo de 'n'
# Come√ßamos em n=2 (pois log(1) = 0 e n>1 para n^log(n))
# Vamos at√© n=50. As fun√ß√µes crescem muito r√°pido!
n = np.linspace(2, 50, 100)

# --- Fun√ß√µes ---

# a) A = n¬≥, B = n log n
A_a = n**3
B_a = n * np.log(n)

# b) A = n log n, B = n^(log n)
A_b = n * np.log(n)
B_b = n**(np.log(n))

# c) A = log(n^k), B = n^(log n)
# Vamos assumir k=2 (qualquer constante k > 0 ter√° o mesmo comportamento assint√≥tico)
k = 2
A_c = k * np.log(n)
B_c = n**(np.log(n))


# --- 2. Criar os 3 Quadros (Gr√°ficos) ---

# Cria a figura e 3 subplots (3 linhas, 1 coluna)
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 18))

# Define um t√≠tulo geral para a figura
fig.suptitle('Compara√ß√£o de Curvas de Complexidade (A vs B)', fontsize=16)

# --- Quadro (a): n¬≥ vs n log n ---
ax1.plot(n, A_a, label='A = $n^3$', color='blue', linewidth=2)
ax1.plot(n, B_a, label='B = $n \log n$', color='red', linestyle='--')
ax1.set_title('a) $A = n^3$  (cresce mais r√°pido $\omega$)  vs  $B = n \log n$')
ax1.set_ylabel('Crescimento (Escala Logar√≠tmica)')
ax1.set_yscale('log') # Essencial usar escala logar√≠tmica!
ax1.legend()
ax1.grid(True)

# --- Quadro (b): n log n vs n^(log n) ---
ax2.plot(n, A_b, label='A = $n \log n$', color='blue', linewidth=2)
ax2.plot(n, B_b, label='B = $n^{\log n}$', color='red', linestyle='--')
ax2.set_title('b) $A = n \log n$  (cresce mais devagar $o$)  vs  $B = n^{\log n}$')
ax2.set_ylabel('Crescimento (Escala Logar√≠tmica)')
ax2.set_yscale('log')
ax2.legend()
ax2.grid(True)

# --- Quadro (c): k*log(n) vs n^(log n) ---
ax3.plot(n, A_c, label=f'A = $k \log n$ (k={k})', color='blue', linewidth=2)
ax3.plot(n, B_c, label='B = $n^{\log n}$', color='red', linestyle='--')
ax3.set_title('c) $A = k \\log n$  (cresce mais devagar $o$)  vs  $B = n^{\\log n}$')
ax3.set_xlabel('Tamanho da Entrada (n)')
ax3.set_ylabel('Crescimento (Escala Logar√≠tmica)')
ax3.set_yscale('log')
ax3.legend()
ax3.grid(True)

# Ajusta o layout para evitar sobreposi√ß√£o
plt.tight_layout(rect=[0, 0.03, 1, 0.96])

# Mostra o gr√°fico
plt.show()
```
